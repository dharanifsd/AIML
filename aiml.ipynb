```python
data=pd.read_csv('/fault.csv')
print("Shape of the Dataset: ", data.shape) # number of rows and columns in our dataset
print("\n\n", data.columns) # columns/features in our Dataset
```

```python
data.tail()
```

```python
# Determining number of fraud cases in Dataset

Fault= data[data['Output (S)'] == 1]
without_fault= data[data['Output (S)'] == 0]
print('Fault Cases: {}'.format(len(data[data['Output (S)'] == 1])))
print('without fault cases: {}'.format(len(data[data['Output (S)'] == 0])))
```

```python
# Spliting the data into features (X) and the target variable (y)
X = data.drop('Output (S)', axis=1)  # Features
y = data['Output (S)']  # Target variable
```

```python
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

With
Naive bayes classifier model

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predicting the labels for the test data
y_pred = gnb.predict(X_test)
```

Accuracy calculation

```python
from sklearn.metrics import accuracy_score
#print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
print("Gaussian Naive Bayes model accuracy(in %):", accuracy_score(y_test, y_pred)*100)
```

```python
#precision score in average value

from sklearn.metrics import precision_score

# Calculate the average precision score
average_precision = precision_score(y_test, y_pred, average="macro")

# Print the average precision score
print("Average precision score:", average_precision)
```

```python
#recall score in average

from sklearn.metrics import recall_score

average_recall = recall_score(y_test, y_pred, average="macro")

print("Average recall score:", average_recall)
```

```python
import numpy as np
from sklearn.metrics import f1_score

# f1 score for each class
f1_scores = f1_score(y_test, y_pred, average=None)

# weighted average of the f1 scores
weighted_f1 = np.average(f1_scores, weights=np.bincount(y_test))

# weighted f1 score
print("F1 score:", weighted_f1)
```

```python
# classification report in tabular column

print(classification_report(y_test, y_pred, target_names=['without_fault', 'Fault']))
```
